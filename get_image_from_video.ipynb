```python
import cv2
import os
video_path = "video/IMG_5384.MP4"
output_folder = "image"
```

```python
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Mở video
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print("Không thể mở video.")
else:
    frame_count = 0
    while True:
        # Đọc từng frame của video
        ret, frame = cap.read()
        if not ret:
            break  # Thoát khi hết video

        # Lưu từng frame vào folder 'image' với định dạng 'frame_x.jpg'
        frame_filename = os.path.join(output_folder, f"frame_{frame_count}.jpg")
        cv2.imwrite(frame_filename, frame)
        
        // print(f"Đã lưu {frame_filename}")
        frame_count += 1

    print("Hoàn tất cắt ảnh từ video.")
    cap.release()
```

```python
import torch

# Kiểm tra nếu có GPU khả dụng
if torch.cuda.is_available():
    print("Có thể sử dụng GPU!")
    print("Số lượng GPU khả dụng:", torch.cuda.device_count())
    print("Tên GPU:", torch.cuda.get_device_name(0))
else:
    print("Không thể sử dụng GPU, chỉ có CPU.")
```

```python
import tensorflow as tf

# Kiểm tra xem có GPU khả dụng hay không
if len(tf.config.experimental.list_physical_devices('GPU')) > 0:
    print("Có thể sử dụng GPU!")
    gpus = tf.config.experimental.list_physical_devices('GPU')
    for gpu in gpus:
        print(f"Tên GPU: {gpu.name}")
else:
    print("Không thể sử dụng GPU, chỉ có CPU.")
```
